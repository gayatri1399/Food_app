```python
import json
import sys
class Admin:
    def __init__(self):
        self.food_item={}
        self.food_id=len(self.food_item)+1
        
    def new_food_items(self):
        self.name=input('Enter food name :')
        self.quantity=(input("Enter food quantity you want to buy :"))
        self.price=int(input('Enter food price :'))
        self.discount=int(input('Enter food discount :'))
        self.stock=int(input('Enter food stock in the restaurant :'))
        
        self.item={'Food name':self.name,'Food quantity':self.quantity,'Food price':self.price,'Food discount':self.discount,'Food stock':self.stock}
        
        self.food_id=len(self.food_item)+1
        self.food_item[self.food_id]=self.item
        with open('Food_item.json','w') as f:
            json.dump(self.food_item,f,indent=4)
            return self.food_item
     
    def edit_food_item(self):  
        food_id=int(input('Enter the id you want to update :'))
        if (food_id)<=len(self.food_item):
            food_name=input('Enter the thing you want to update :')
            value=input('Enter the value you want to replace with old value :')
            with open("Food_item.json","r") as f:
                self.food_item=json.load(f)
            self.food_item[str(food_id)][food_name]=value
            with open('Food_item.json','w') as f:
                json.dump(self.food_item,f,indent=4)
            return self.food_item
        return self.food_item
       
    
    def load_food_item(self):
        with open("Food_item.json","r") as f:
            self.food_item=json.load(f)
        return self.food_item
    
    def remove_food_item(self):
        food_item=int(input('Enter the key you want to delete :'))
        if (food_item)<len(self.food_item):
            with open("Food_item.json","r") as f:
                self.food_item=json.load(f)
        del self.food_item[str(food_item)]
        #print('Remaining food item is :')
        with open('Food_item.json','w') as f:
            json.dump(self.food_item,f,indent=4)
        return self.food_item
        
        
    
c=Admin()

while True:
    print('*'*100)
    print("Press 1 -- To add new food items")
    print("Press 2 -- To edit food items")
    print("Press 3 -- To load(view) food items")
    print("Press 4 -- To remove food items")
    print("Press 5 -- FOR EXIT")
    print("*************************************************************************************")
    option = int(input("Choose your Option for Transaction :"))
    if option==1:
        c.new_food_items()
        print('*'*100)
    elif option==2:
        c.edit_food_item()
        print('*'*100)
    elif option==3:
        print('*'*100)
        print(c.load_food_item())
    elif option==4:
        c.remove_food_item()
        print('*'*100)
    elif option==5:
        print('Thanks for visiting! Visit Again')
        sys.exit()
    else:
        print("Please enter valid input")
```

```python
import json
import sys
class User:
    def __init__(self):
        self.register={}
        self.register_id=len(self.register)+1
        self.em_pass={}
        self.orders={}
    def registeration(self):
        self.name=input('Enter your name :')
        self.ph_no=int(input('Enter phone number :'))
        self.email=input('Enter email address :')
        self.address=input('Enter your address :')
        self.password=input('Enter your password :')
        self.register_details={'Name':self.name,'Ph number':self.ph_no,'Email':self.email,'Address':self.address,'Password':self.password}
        
        self.register_id=len(self.register)+1
        self.register[self.register_id]=self.register_details
        self.em_pass[self.email]=self.password
        with open('registeration_details.json','w') as f:
            json.dump(self.register,f,indent=4)
        with open('email_details.json','w') as f:
            json.dump(self.em_pass,f,indent=4)

        return self.register
    
    def load_register_details(self):
        with open("registeration_details.json","r") as f:
            self.register=json.load(f)
        return self.register
    
    def log_in(self):
        email_id=input('Enter the email address to login :')
        for i in self.register.values():
            if any(i['Email'])==email_id:
                break 
        Password=input('Please enter your password :')
        with open("email_details.json","r") as f:
            self.em_pass=json.load(f)
        for i in self.em_pass:
            if email_id in i:
                #break
                if self.em_pass[email_id]==Password:
                    print('Login successful! Welcome to resturant!')
                    break
        else:
            print('You are new user. Please register')
        return ""
            
    def place_order(self):
        print('*'*100)
        print('The menu card is :')
        with open("Food_item.json","r") as f:
            self.food_item=json.load(f)
        menu=self.food_item
        for i in self.food_item:
            food=('{} : {}, ({}),[INR :{}] '.format(str(i),self.food_item[str(i)]['Food name'],self.food_item[str(i)]['Food quantity'],self.food_item[str(i)]['Food price']))
            print(food)
        a=list((input('Enter the food id you need to buy :')).split())
       
        for i in a:
            if i in menu.keys():
                self.order=('{}, ({}),[INR :{}] '.format(menu[str(i)]['Food name'],menu[str(i)]['Food quantity'],menu[str(i)]['Food price']))
                self.orders[i]=self.order
        with open('Order_history.json','w') as f:
            json.dump(self.orders,f,indent=4)
        return ''
                
    def order_history(self):
        print('*'*100)
        print('The food items you ordered are :')
        with open("Order_history.json","r") as f:
            self.orders=json.load(f)
        return self.orders
    
    def update_profile(self):
        with open("registeration_details.json","r") as f:
            self.register=json.load(f)
        update=input('Enter what you want to update :')
        value=input('Enter the value you want to update :')
        self.register['1'][update]=value
        with open('registeration_details.json','w') as f:
            json.dump(self.register,f,indent=4)
        print('The updated profile is :',self.register)
        return ''
            
obj=User()
print(obj.registeration())
print(obj.log_in())
print(obj.log_in())
while True:
    print('*'*100)
    print("Press 1 -- To place order ")
    print("Press 2 -- For order history ")
    print("Press 3 -- to update profile")
    print("Press 4 -- FOR EXIT")
    print("*************************************************************************************")
    option = int(input("Choose your Option for Transaction :"))
    if option==1:
        obj.place_order()
    elif option==2:
        print(obj.order_history())
    elif option==3:
        obj.update_profile()
    elif option==4:
        print('Thanks for visiting')
        sys.exit()
    else:
        print("Please enter valid input")
```
